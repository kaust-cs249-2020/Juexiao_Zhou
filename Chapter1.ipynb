```python
def MaxMap(freqMap):
    return max(freqMap.values())

def FrequentWords(seq,k=1):
    freqMap={}
    out=[]
    for i in range(len(seq)-k+1):
        try:
            freqMap[seq[i:i+k]]+=1
        except:
            freqMap[seq[i:i+k]]=1
    maxFrequent=MaxMap(freqMap)
    for key in freqMap:
        if freqMap[key]==maxFrequent:
            out.append(key)
    return out

FrequentWords('ACGTTGCATGTCGCATGATGCATGAGAGCT',4)
```

```python
def FrequentTable(seq,k=1):
    freqMap={}
    for i in range(len(seq)-k+1):
        try:
            freqMap[seq[i:i+k]]+=1
        except:
            freqMap[seq[i:i+k]]=1
    return freqMap

def FindClumps(Text,k,L,t):
    patterns=[]
    for i in range(len(Text)-L+1):
        window=Text[i:i+L]
        fmap=FrequentTable(window,k)
        for key in fmap.keys():
            if fmap[key]>=t:
                patterns.append(key)
    return set(patterns)

FindClumps('CGGACTCGACAGATGTGAAGAACGACAATGTGAAGACTCGACACGACAGAGTGAAGAGAAGAGGAAACATTGTAA',5,50,4)
```

## 1.7 Peculiar Statistics of the Forward and Reverse Half-Strands

```python
def Skew(seq):
    count=0
    out=[count]
    for x in seq:
        if x=='C':
            count-=1
        elif x=='G':
            count+=1
        out.append(count)
    return out

skew('CATGGGCATCGGCCATACGCC')
```

```python
' '.join([str(x) for x in skew('GAGCCACCGCGATA')])
```

```python
def Minimum_Skew(seq):
    skew=Skew(seq)
    min_skew=min(skew)
    out=[]
    for i in range(len(skew)):
        if skew[i]==min_skew:
            out.append(i)
    return out

Minimum_Skew('TAAAGACTGCCGAGAGGCCAACACGAGTGCTAGAACGAGGGGCGTAAACGCGGGTCCGAT')
```
