## 3.8 From Euler's Theorem to an Algorithm for Finding Eulerian Cycles

```python
import random

def EulerianCycle(Graph):
    tmp_Graph=Graph
    current=random.choice(list(tmp_Graph.keys()))
    Cycle=[]
    n_edges=0
    for key in Graph.keys():
        n_edges+=len(Graph[key]) # store the total# edges
    while tmp_Graph[current]!=[]:
        Cycle.append(current)
        tmp=current
        current=random.choice(tmp_Graph[current])
        tmp_Graph[tmp].remove(current)
        n_edges-=1
    Cycle.append(current) # form a cycle Cycle by randomly walking in Graph
    
    while n_edges!=0:
        newStarts=[x for x in Cycle if tmp_Graph[x]!=[]]
        newStart=random.choice(newStarts)
        _Cycle=Cycle[Cycle.index(newStart):len(Cycle)-1]+Cycle[0:Cycle.index(newStart)+1]
        current=newStart
        while tmp_Graph[current]!=[]:
            tmp=current
            current=random.choice(tmp_Graph[current])
            tmp_Graph[tmp].remove(current)
            _Cycle.append(current)
            n_edges-=1
        Cycle=_Cycle
        
    return '->'.join(Cycle)
    
    
Graph={'0':['3'],
      '1':['0'],
      '2':['1','6'],
      '3':['2'],
      '4':['2'],
      '5':['4'],
      '6':['5','8'],
      '7':['9'],
      '8':['7'],
      '9':['6']}

EulerianCycle(Graph)
```

```python
# cite from https://github.com/xuwd11/Coursera-Bioinformatics/blob/master/22_UniversalCircularString.py

def de_bruijn(k, n):
    """
    De Bruijn sequence for alphabet k
    and subsequences of length n.
    """
    try:
        # let's see if k can be cast to an integer;
        # if so, make our alphabet a list
        _ = int(k)
        alphabet = list(map(str, range(k)))

    except (ValueError, TypeError):
        alphabet = k
        k = len(k)

    a = [0] * k * n
    sequence = []

    def db(t, p):
        if t > n:
            if n % p == 0:
                sequence.extend(a[1:p + 1])
        else:
            a[t] = a[t - p]
            db(t + 1, p)
            for j in range(a[t - p] + 1, k):
                a[t] = j
                db(t + 1, t)
    db(1, 1)
    return "".join(alphabet[i] for i in sequence)

de_bruijn(2, 3)
```

### 016. String Reconstruction from Read-Pairs Problem

```python
import random
from collections import Counter

def Prefix_Paired(kd_mers): # CCGG|AATT -> CCG|AAT
    kd_mers=kd_mers.split('|')
    return '|'.join([kd_mers[0][:-1],kd_mers[1][:-1]])

def Suffix_Paired(kd_mers): # CCGG|AATT -> CGG|ATT
    kd_mers=kd_mers.split('|')
    return '|'.join([kd_mers[0][1::],kd_mers[1][1::]])

def DeBruijn_Graph_from_Paired_kmers_Problem(paired_kmers):
    out={}
    for paired_kmer in paired_kmers:
        try:
            out[Prefix_Paired(paired_kmer)].append(Suffix_Paired(paired_kmer))
        except:
            out[Prefix_Paired(paired_kmer)]=[Suffix_Paired(paired_kmer)]
    out=dict((k,out[k]) for k in sorted(out.keys()))
    
    #for key in out:
    #    print(key+' -> ',end='')
    #    print(','.join(out[key]))
    return out
        
def unbalanced_nodes(Graph):
    nodes=[]
    out_degree={key:len(Graph[key]) for key in Graph.keys()}
    _in=[]
    for x in Graph.values():
        _in+=x
    in_degree=Counter(_in)
    for key in Graph.keys():
        nodes.append(key)
        nodes+=Graph[key]
    nodes=set(nodes)
    #print(in_degree,out_degree)
    to_out=None
    to_in=None
    for node in nodes:
        if node in in_degree.keys() and node in out_degree.keys():
            if in_degree[node]>out_degree[node]:
                to_out=node
            elif in_degree[node]<out_degree[node]:
                to_in=node
        elif node not in in_degree.keys() and node in out_degree.keys():
            to_in=node
        elif node not in out_degree.keys() and node in in_degree.keys():
            to_out=node
    return to_in,to_out

def EulerianCycle(Graph):
    tmp_Graph=Graph
    current=random.choice(list(tmp_Graph.keys()))
    Cycle=[]
    n_edges=0
    for key in Graph.keys():
        n_edges+=len(Graph[key]) # store the total# edges
    while tmp_Graph[current]!=[]:
        Cycle.append(current)
        tmp=current
        current=random.choice(tmp_Graph[current])
        tmp_Graph[tmp].remove(current)
        n_edges-=1
    Cycle.append(current) # form a cycle Cycle by randomly walking in Graph
    
    while n_edges!=0:
        newStarts=[x for x in Cycle if tmp_Graph[x]!=[]]
        newStart=random.choice(newStarts)
        _Cycle=Cycle[Cycle.index(newStart):len(Cycle)-1]+Cycle[0:Cycle.index(newStart)+1]
        current=newStart
        while tmp_Graph[current]!=[]:
            tmp=current
            current=random.choice(tmp_Graph[current])
            tmp_Graph[tmp].remove(current)
            _Cycle.append(current)
            n_edges-=1
        Cycle=_Cycle
        
    return Cycle

def EulerianPath(Cycle,to_in,to_out): # find the eulerian path in the paired_reads DeBruijn_Graph
    _Cycle=Cycle[Cycle.index(to_out)+1:len(Cycle)-1]+Cycle[0:Cycle.index(to_out)+1]
    return _Cycle
    
def StringSpelledByGappedPatterns(GappedPatterns, k, d): # ['xxx|xxx','xxx|xxx'] -> final seqeuence xxxxx
    GappedPatterns=[x.split('|') for x in GappedPatterns]
    FirstPatterns=[x[0] for x in GappedPatterns]
    SecondPatterns=[x[1] for x in GappedPatterns]
    #print(FirstPatterns,SecondPatterns)
    PrefixString=StringSpelledByPatterns(FirstPatterns)
    SuffixString=StringSpelledByPatterns(SecondPatterns)
    for i in range(k+d,len(PrefixString)):
        if PrefixString[i]!=SuffixString[i-k-d]:
            return "there is no string spelled by the gapped patterns"
    return ''.join([PrefixString,SuffixString[-k-d::]])

def StringReconstruction(Patterns,k,d):
    db=DeBruijn_Graph_from_Paired_kmers_Problem(Patterns)
    to_in,to_out=unbalanced_nodes(db)
    try:
        db[to_out].append(to_in)
    except:
        db[to_out]=[to_in]
    Cycle=EulerianCycle(db)
    path=EulerianPath(Cycle,to_in,to_out)
    #print(Cycle,path)
    text=StringSpelledByGappedPatterns(path,k,d)
    return text

Paired_Reads='GAGA|TTGA \
TCGT|GATG \
CGTG|ATGT \
TGGT|TGAG \
GTGA|TGTT \
GTGG|GTGA \
TGAG|GTTG \
GGTC|GAGA \
GTCG|AGAT'.split(' ')
k=4
d=2

StringReconstruction(Paired_Reads,k,d)
```
