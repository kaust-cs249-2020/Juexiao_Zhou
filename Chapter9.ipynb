## 9.3 Herding Patterns into a Trie

### Trie Construction Problem

```python
def TrieConstruction(Patterns):
    Trie={0:{}} # new graph
    nodeCount=0
    for Pattern in Patterns:
        currentNode=0
        for i in range(len(Pattern)):
            currentSymbol=Pattern[i]
            if currentSymbol in Trie[currentNode].values():
                currentNode={v:k for k,v in Trie[currentNode].items()}[currentSymbol]
            else:
                oldNode=currentNode
                nodeCount+=1
                newNode=nodeCount
                Trie[oldNode][newNode]=currentSymbol
                if newNode not in Trie.keys():
                    Trie[newNode]={}
                currentNode=newNode
                #print(Trie)
    return Trie

def FormatTrie(Trie):
    for x in Trie:
        if len(Trie[x])!=0:
            for y in Trie[x]:
                print('%s->%s:%s'%(str(x),str(y),Trie[x][y]))
                
Patterns=['ATAGA','ATC','GAT']
Trie=TrieConstruction(Patterns)
FormatTrie(Trie)
```

### PrefixTrieMatching

```python
def DFSFindPathInTrie(i,j,Trie):
    stack = [(i,[i])]
    visited = set()
    while stack:
        (vertex, path) = stack.pop()
        if vertex not in visited:
            if vertex == j:
                return path
            visited.add(vertex)
            for neighbor in Trie[vertex].keys():
                stack.append((neighbor, path + [neighbor]))

def TrieConstruction(Patterns):
    Trie={0:{}} # new graph
    nodeCount=0
    for Pattern in Patterns:
        currentNode=0
        for i in range(len(Pattern)):
            currentSymbol=Pattern[i]
            if currentSymbol in Trie[currentNode].values():
                currentNode={v:k for k,v in Trie[currentNode].items()}[currentSymbol]
            else:
                oldNode=currentNode
                nodeCount+=1
                newNode=nodeCount
                Trie[oldNode][newNode]=currentSymbol
                if newNode not in Trie.keys():
                    Trie[newNode]={}
                currentNode=newNode
                #print(Trie)
    return Trie

def PrefixTrieMatching(Text, Trie):
    symbol_pos=0
    symbol=Text[symbol_pos]
    v=0
    while True:
        if len(Trie[v])==0: # a leaf
            return DFSFindPathInTrie(0,v,Trie)
        elif symbol in Trie[v].values():
            w={value:key for key,value in Trie[v].items()}[symbol]
            symbol_pos+=1
            try:
                symbol=Text[symbol_pos]
            except:
                pass
            v=w
            #print(v)
        else:
            return False
        
def TrieMatching(Text, Trie):
    for i in range(len(Text)):
        if PrefixTrieMatching(Text[i::], Trie)!=False:
            print(i,end=' ')
        
Text='AATCGGGTTCAATCGGGGT'
Patterns=['ATCG','GGGT']
Trie=TrieConstruction(Patterns)
#PrefixTrieMatching('ATCG',Trie)
TrieMatching(Text, Trie)
```

```python
## modified from 3.14

from collections import Counter

def MaximalNonBranchingPaths_Growth(Graph,start):
    def OutDegree(Graph,node):
        if node not in Graph.keys():
            return 0
        else:
            return len(Graph[node])
    
    NonBranchingPaths=[]
    Stack=[[start]]
    while len(Stack)!=0:
    #for i in range(1):
        #print(Stack)
        path=Stack.pop(0)
        last_node=path[-1]
        if OutDegree(Graph,last_node)==0:
            NonBranchingPaths.append(path)
        elif OutDegree(Graph,last_node)==1:
            next_node=Graph[last_node][0]
            Stack.append(path+[next_node])
        elif OutDegree(Graph,last_node)>1:
            NonBranchingPaths.append(path)
            if last_node!=start:
                for next_node in Graph[last_node]:
                    Stack.append([last_node,next_node])
            else:
                for next_node in Graph[last_node]:
                    Stack.append(path+[next_node])
            #print(Stack)
    return [x for x in NonBranchingPaths if len(x)>2]

class Graph():
    def __init__(self):
        self.nodes=[]
        self.nodes_label={}
        self.nodes_num=0
        
        self.edges={}
        self.edges_label={}
        self.edges_position={}
        self.edges_num=0
        self.edges_length={}
        
    def addNode(self,node1):
        self.nodes.append(node1)
        self.nodes_num+=1
        
    def addEdge(self,node1,node2):
        try:
            self.edges[node1].append(node2)
        except:
            self.edges[node1]=[node2]
        self.edges_num+=1
        
    def addNodeLabel(self,node1,label):
        self.nodes_label[node1]=label
        
    def addEdgeLabel(self,edge1,label):
        self.edges_label[edge1]=label
        
    def addEdgePosition(self,edge1,position):
        self.edges_position[edge1]=position
        
    def addEdgeLength(self,edge1,length):
        self.edges_length[edge1]=length
        
    def delNode(self,node1):
        self.nodes.remove(node1)
        if node1 in self.nodes_label.keys():
            del self.nodes_label[node1]
        self.nodes_num-=1
        
    def delEdge(self,edge1):
        self.edges[edge1[0]].remove(edge1[1])
        #del self.edges_label[edge1]
        del self.edges_length[edge1]
        if len(self.edges[edge1[0]])==0:
            del self.edges[edge1[0]]
        self.edges_num-=1
        
    def hasOutgoingEdgesWithLabelFromNode(self,node1,label):
        #print(self.edges)
        try:
            for node2 in self.edges[node1]:
                if self.edges_label[(node1,node2)]==label:
                    return True
        except:
            pass
        return False
        
    def NextNodeInEdgeWithLabelFromNode(self,node1,label):
        for node2 in self.edges[node1]:
            if self.edges_label[(node1,node2)]==label:
                return node2

    def isLeafNode(self,node1):
        if node1 not in self.edges.keys():
            return True
        else:
            return False
        
    def toLinkedList(self):
        return self.edges


def ModifiedTrieConstruction(Text):
    Trie=Graph()
    Trie.addNode(0)
    Trie.addNodeLabel(0,'root')
    nodeCount=0
    for i in range(len(Text)):
        print(i,'/',len(Text))
        currentNode=0
        for j in range(i,len(Text)):
            currentSymbol=Text[j]
            if Trie.hasOutgoingEdgesWithLabelFromNode(currentNode,currentSymbol):
                currentNode=Trie.NextNodeInEdgeWithLabelFromNode(currentNode,currentSymbol)
            else:
                oldNode=currentNode
                nodeCount+=1
                newNode=nodeCount
                Trie.addNode(newNode)
                Trie.addEdge(oldNode,newNode)
                Trie.addEdgeLabel((oldNode,newNode),currentSymbol)
                Trie.addEdgePosition((oldNode,newNode),j)
                Trie.addEdgeLength((oldNode,newNode),1)
                currentNode=newNode
        if Trie.isLeafNode(currentNode):
            Trie.addNodeLabel(currentNode,i)
    return Trie

def ModifiedSuffixTreeConstruction(Text):
    Trie=ModifiedTrieConstruction(Text)
    NonBranchingPaths=MaximalNonBranchingPaths_Growth(Trie.toLinkedList(),0)
    count=0
    print(NonBranchingPaths)
    for nonBranchingPath in NonBranchingPaths:
        print(count,'/',len(NonBranchingPaths))
        count+=1
        for i in range(len(nonBranchingPath)-1):
            Trie.delEdge((nonBranchingPath[i],nonBranchingPath[i+1]))
        # because of here, only Trie.edges is up-to-date
        #for i in range(1,len(nonBranchingPath)): # no need to run this, will be much quicker
        #    Trie.delNode(nonBranchingPath[i])
        Trie.addEdge(nonBranchingPath[0],nonBranchingPath[-1])
        Trie.addEdgePosition((nonBranchingPath[0],nonBranchingPath[-1]),Trie.edges_position[(nonBranchingPath[0],nonBranchingPath[1])])
        Trie.addEdgeLength((nonBranchingPath[0],nonBranchingPath[-1]),len(nonBranchingPath)-1)
        Trie.addEdgeLabel((nonBranchingPath[0],nonBranchingPath[-1]),''.join([Trie.edges_label[(nonBranchingPath[i],nonBranchingPath[i+1])] for i in range(len(nonBranchingPath)-1)]))
    return Trie

def SuffixTree(Text):
    Trie=ModifiedSuffixTreeConstruction(Text)
    
    # sort edges in order "$abc...": sorted([]) is good
    for node1 in Trie.edges.keys():
        to_sort_node2s={} # {node2:edge_label,node2:edge_label}
        node2s=Trie.edges[node1]
        #print('before',Trie.edges[node1])
        for node2 in node2s:
            to_sort_node2s[node2]=Trie.edges_label[(node1,node2)]
        sorted_node2s=sorted(to_sort_node2s.items(), key=lambda x:x[1])
        Trie.edges[node1]=[x[0] for x in sorted_node2s]
        #print('after',Trie.edges[node1])
        #print(sorted_node2s)
    
    return Trie

def SuffixArray(Text):
    suffixTree=SuffixTree(Text)
    Array=[] # should find the position from root->leaf
    
    def DFSFindPathInTrie(i,j,Trie):
        stack = [(i,[i])]
        visited = set()
        while stack:
            (vertex, path) = stack.pop()
            if vertex not in visited:
                if vertex == j:
                    return path
                visited.add(vertex)
                for neighbor in Trie.edges[vertex]:
                    if neighbor in Trie.edges.keys():
                        stack.append((neighbor, path + [neighbor]))
                    else:
                        if neighbor==j:
                            stack.append((neighbor, path + [neighbor]))
                        
    def Preorder(Trie,node1):
        if node1 in Trie.edges.keys():
            for node2 in Trie.edges[node1]:
                if node2 not in Trie.edges.keys(): # node2 is the leaf
                    path=DFSFindPathInTrie(0,node2,suffixTree)
                    starting_position=Trie.edges_position[(path[-2],path[-1])] # here the starting position only record the position of last edge
                    for i in range(len(path)-2): # then we eliminate the edge length of previous edges in the path fro root to leaf
                        starting_position-=Trie.edges_length[(path[i],path[i+1])]
                    print(starting_position)
                    Array.append(starting_position)
                Preorder(Trie,node2)
    Preorder(suffixTree,0)
    return Array
    
Text='AACGATAGCGGTAGA$'
suffixArray=SuffixArray(Text)
print(suffixArray)
```

```python
def BWT(Text):
    Cyclic_Rotations=[Text]
    for i in range(len(Text)-1):
        tmp=Cyclic_Rotations[-1]
        Cyclic_Rotations.append(tmp[-1]+tmp[:-1])
    M=sorted(Cyclic_Rotations)
    out=''.join([x[-1] for x in M])
    #print(Cyclic_Rotations)
    return out

Text='GCGTGCCTGGTCA$'
BWT(Text)
```

```python
import numpy as np

def getLastToFirst(transform):
    firstColumn=sorted(list(transform))
    FirstColumn=[]
    lastColumn=list(transform)
    LastColumn=[]
    def encode_relation(firstColumn,lastColumn):
        first2last={}
        last2first={}
        
        first_letter_count={}
        last_letter_count={}
        
        for i in range(len(firstColumn)):
            try:
                first_node_count=first_letter_count[firstColumn[i]]
            except:
                first_letter_count[firstColumn[i]]=1
                first_node_count=first_letter_count[firstColumn[i]]
            try:
                last_node_count=last_letter_count[lastColumn[i]]
            except:
                last_letter_count[lastColumn[i]]=1
                last_node_count=last_letter_count[lastColumn[i]]
            first2last['{}_{}'.format(firstColumn[i],first_letter_count[firstColumn[i]])]='{}_{}'.format(lastColumn[i],last_letter_count[lastColumn[i]])
            last2first['{}_{}'.format(lastColumn[i],last_letter_count[lastColumn[i]])]='{}_{}'.format(firstColumn[i],first_letter_count[firstColumn[i]])
            FirstColumn.append('{}_{}'.format(firstColumn[i],first_letter_count[firstColumn[i]]))
            LastColumn.append('{}_{}'.format(lastColumn[i],last_letter_count[lastColumn[i]]))
            first_letter_count[firstColumn[i]]+=1
            last_letter_count[lastColumn[i]]+=1
        return first2last,last2first
    first2last,last2first=encode_relation(firstColumn,lastColumn)
    LastToFirst=[]
    for x in LastColumn:
        LastToFirst.append(FirstColumn.index(x))
    #print(FirstColumn,LastColumn)
    return LastToFirst

def FirstOccurrence(symbol,LastColumn):
    firstColumn=sorted(list(LastColumn))
    for i in range(len(firstColumn)):
        if firstColumn[i]==symbol:
            return i
        
def count(i,symbol,List):
    c=0
    for j in range(i):
        if List[j]==symbol:
            c+=1
    return c

def SuffixArray(text):
    suffixes = []
    for i in range(len(text)):
        suffixes.append((text[i:], i))
    suffixes.sort()
    return [x[1] for x in suffixes]

def Pattern2Seeds(Pattern,d):
    #print(len(Pattern)/(d+1),np.floor(len(Pattern)/(d+1)))
    k=len(Pattern)//(d+1)
    break_idx=list(range(0,len(Pattern)-k+1,k))+[len(Pattern)]
    seeds=[[Pattern[break_idx[i-1]:break_idx[i]],break_idx[i-1]] for i in range(1,len(break_idx))]
    return seeds

def BetterBWMatchingPosition(LastColumn,seed,LastToFirst,suffixArray):
    top=0
    bottom=len(LastColumn)-1
    while top<=bottom:
        if len(seed)!=0:
            symbol=seed[-1]
            seed=seed[:-1]
            if symbol in LastColumn[top:bottom+1]:
                top=FirstOccurrence(symbol,LastColumn)+count(top,symbol,LastColumn)
                bottom=FirstOccurrence(symbol,LastColumn)+count(bottom + 1,symbol,LastColumn)-1
            else:
                return []
        else:
            return [suffixArray[i] for i in range(top,bottom+1)]
    return []

def ApproximatePatternPosition(Text,LastColumn,Pattern,LastToFirst,suffixArray,d):
    pattern_positions=[]
    seeds=Pattern2Seeds(Pattern,d)
    for seed,break_pos in seeds: # break pattern into small parts
        print(seed,break_pos)
        tmp_positions=BetterBWMatchingPosition(LastColumn,seed,LastToFirst,suffixArray) # matching small parts
        print(tmp_positions)
        for tmp_pos in tmp_positions: # for each candidate position
            pattern_pos=tmp_pos-break_pos
            flag=1
            if pattern_pos<0: # if the matching is before 
                flag=0
            if pattern_pos+len(Pattern)>len(Text): # match after
                flag=0
            ## then test if is approximate matching
            if flag==1:
                mismatch_score=0
                for i in range(len(Pattern)):
                    if Pattern[i]!=Text[pattern_pos+i]:
                        mismatch_score+=1
                if mismatch_score<=d: # calculate the mismatch score
                    pattern_positions.append(pattern_pos)
    return list(set(pattern_positions))
        
def BWT(Text):
    Cyclic_Rotations=[Text]
    for i in range(len(Text)-1):
        tmp=Cyclic_Rotations[-1]
        Cyclic_Rotations.append(tmp[-1]+tmp[:-1])
    M=sorted(Cyclic_Rotations)
    out=''.join([x[-1] for x in M])
    #print(Cyclic_Rotations)
    return out

def Multiple_Approximate_Pattern_Matching(Text,Patterns,d):
    LastColumn=BWT(Text)
    LastToFirst=getLastToFirst(LastColumn)
    suffixArray=SuffixArray(Text)
    print(suffixArray)
    positions=[]
    c=0
    for Pattern in Patterns:
        print(c,'/',len(Patterns))
        c+=1
        position=ApproximatePatternPosition(Text,LastColumn,Pattern,LastToFirst,suffixArray,d)
        positions+=position
    return sorted(positions)

Text='ACATGCTACTTT$'
#Patterns=['ATT']
Patterns='ATT GCC GCTA TATT'.split(' ')
d=1 
print(Multiple_Approximate_Pattern_Matching(Text,Patterns,d))

#Patterns='ACC CCG CAG'.split(' ')
#for Pattern in Patterns:
#    print(BetterBWMatching(LastColumn,Pattern,LastToFirst),end=' ')
```

### LCP(Text)

```python
def LCP(Text):
    suffixArray=SuffixArray(Text)
    sortedSuffixes=SortedSuffixes(Text, suffixArray)
    LCP_Array=[0]
    def score(seq1,seq2):
        length=min(len(seq1),len(seq2))
        s=0
        for ii in range(length):
            if seq1[ii]==seq2[ii]:
                s+=1
            else:
                break
        return s
        
    for i in range(len(sortedSuffixes)-1):
        seq1=sortedSuffixes[i]
        seq2=sortedSuffixes[i+1]
        LCP_Array.append(score(seq1,seq2))
    return LCP_Array

Text='panamabananas$'
LCP_Array=LCP(Text)
print(LCP_Array)
```
